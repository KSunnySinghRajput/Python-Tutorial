```python
import os

def add_path(path):  
        envdir_list=os.environ["PATH"].replace(os.pathsep+os.pathsep, os.pathsep).split(os.pathsep)
        if path in envdir_list:
            print("Given Path exist")
        else:
            if os.path.isfile(path) and os.access(path, os.X_OK):
                if os.path.exists(path):
                    os.environ["PATH"]+=os.pathsep+os.path.normcase(path)  # / --> \\
                    print("Success.")
                elif(os.path.exists(os.curdir+path)):
                    os.environ["PATH"]+=os.pathsep+os.path.join(os.getcwd(), path)
                    print("Success.")
            else:
                print('Given Path Not exist!!!!')
```

```python
add_path('./C')
```
