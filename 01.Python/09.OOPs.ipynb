# Class Methods & Static Methods

```python
class Employee:
    no_of_leaves = 30       #Class Variable
    def __init__(self, name, age, salary):
        self.name = name
        self.age = age
        self.salary = salary

    def printdetails(self):
        return f"The name is {self.name}, age is {self.age} and salary is {self.salary}"
    
    #If you want to change the value of class variables using function
    #Then you should use class method
    @classmethod
    def changeleaves(cls, newleaves):
        cls.no_of_leaves = newleaves

    #If you want to put only a simple method that takes neither class variables nor instance variables
    # Then you should use static method (If will only print a message that you want to display)    
    @staticmethod
    def printstatic():
        print("This is example of static method")

# Employee.no_of_leaves = 50          #Here you can change the no_of_leaves
emp1 = Employee("Shayam", 34, 34000)
print(emp1.printdetails())
emp1.changeleaves(35)
print(Employee.no_of_leaves)
print(emp1.no_of_leaves)
emp1.printstatic()                     #Here I am calling static method
```

# Scopes and Namespaces

```python
def scope_test():
    def do_local():
        spam = "local spam"
    def do_nonlocal():
        nonlocal spam
        spam = "nonlocal spam"
    def do_global():
        global spam
        spam = "global spam"
    spam = "test spam"
    do_local()
    print("After local assignment:", spam)    #10 line o/p
    do_nonlocal()
    print("After nonlocal assignment:", spam)   #6 line o/p
    do_global()
    print("After global assignment:", spam)     #6 line o/p
scope_test()
print("In global scope:", spam)         #9 line o/p
```

## 2.Multiple Inheritance 

```python
#Parent Class
class Father:
    no_of_leaves = 30
    def __init__(self, fname, lname, age, occupation):
        self.fname = fname
        self.lname = lname
        self.age = age
        self.occupation = occupation
    def fdetails(self):
        return f"Name is {self.fname} {self.lname}, age is {self.age} and occupation is {self.occupation}"
    # def fleaves(self):
    #     print(self.fleaves)

#Parent Class
class Mother:
    no_of_leaves = 30
    def __init__(self, fname, lname, age, expertise):
        self.fname = fname
        self.lname = lname
        self.age = age
        self.expertise = expertise
    def mdetails(self):
        return f"Name is {self.fname} {self.lname} and age is {self.age} and expertise is {self.expertise}"
    # def mexpertise(self):
    #     print (self.cooking_expertise)

#Child Class - It Inherits from both above classes
class Child(Father, Mother):
    no_of_leaves = 30
    def cdetails(self):
        return f"Name is {self.fname} {self.lname}, age is {self.age} and he is {self.occupation}"

ramesh = Father("Ramesh","Agarwal", 50, "Business")
print(ramesh.fdetails())
# ramesh.fleaves()

sushma = Mother("Sushma", "Agarwal", 47, "Cooking")
print(sushma.mdetails())

suresh = Child("Suresh","Agarwal", 12, "Student")
print(suresh.cdetails())
print(suresh.no_of_leaves)
```

## 3.Multilevel Inheritance

```python
#Parent Class
class Grandfather:
    no_of_leaves = 10
    def __init__(self, name, age):
        self.name = name
        self.age = age  
    def gfdetails(self):
        return f"Name is {self.name} and age is {self.age}"

class Father(Grandfather):
    no_of_leaves = 15
    def __init__(self, name, age, occupation):
        self.name = name
        self.age = age
        self.occupation = occupation
    def fdetails(self):
        return f"Name is {self.name}, age is {self.age} and occupation is {self.occupation}"    

class Child(Father):
    no_of_leaves = 20
    def fdetails(self):
        return f"Name is {self.name}, age is {self.age} and occupation is {self.occupation} and my leaves are {self.no_of_leaves}" 
        
ramesh = Grandfather("Ramesh",50)
print(ramesh.gfdetails())

mahesh = Father("Mahesh", 47, "Teacher")
print(mahesh.fdetails())
print(mahesh.no_of_leaves)

suresh = Child("Suresh", 12, "Student")
print(suresh.fdetails())        
# print(suresh.no_of_leaves)
         



    
```

## 4.Super Inheritance

```python
class Employee:
    def __init__(self, fname, lname, age, salary):
        self.fname = fname
        self.lname = lname
        self.age = age
        self.salary = salary
    def edetails(self):
        return f"Name is {self.fname} {self.lname}, age is {self.age} and salary is {self.salary}"


class Programmer(Employee):
    def __init__(self, fname, lname, age, salary, language, role):
        super().__init__(fname, lname, age, salary)
        self.language = language
        self.role = role
    def pdetails(self):
        return f"Name is {self.fname} {self.lname}, age is {self.age}, salary is {self.salary}, language is {self.language} and role is {self.role}"


ramesh = Employee("Ramesh","Agarwal", 35, 50000)
print(ramesh.edetails())
suresh = Programmer("Suresh","Sharma", 35, 45000, "Python", "Python Programmer")
print(suresh.pdetails())
```
